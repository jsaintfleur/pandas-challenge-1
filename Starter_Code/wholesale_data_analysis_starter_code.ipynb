## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

# Make a copy of the original dataframe to preserve the original data
updated_wholesales_df = wholesales_df.copy()

# Create a new column 'subtotal' that calculates the subtotal for each line
updated_wholesales_df['subtotal'] = updated_wholesales_df['unit_price'] * updated_wholesales_df['qty']

# Display the first few rows to check the new column
updated_wholesales_df.head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.


# Define a function to calculate the shipping price based on the total weight (unit weight * qty)
def calculate_shipping_price(row):
    # Calculate the total weight by multiplying unit weight with qty
    total_weight = row['unit_weight'] * row['qty']
    
    # Apply the shipping price logic based on the total weight
    if total_weight > 50:
        return total_weight * 7  # $7 per pound for orders over 50 pounds
    else:
        return total_weight * 10  # $10 per pound for items 50 pounds or under

# Apply the function to create the 'shipping_price' column
updated_wholesales_df['shipping_price'] = updated_wholesales_df.apply(calculate_shipping_price, axis=1)

# Display the first few rows to check the new 'shipping_price' column
updated_wholesales_df.head()


```

```python
# Check if there are any items with a unit weight greater than 50 lbs
items_over_50_lbs = updated_wholesales_df['unit_weight'] > 50

# Display the result (True if there are any, False if not)
items_over_50_lbs.any()
```

```python
# Define a function to calculate the total price including subtotal, shipping price, and a sales tax of 9.25%
def calculate_total_price(row):
    # Calculate the combined price (subtotal + shipping price)
    combined_price = row['subtotal'] + row['shipping_price']
    # Apply the sales tax of 9.25% to the combined price
    total_price = combined_price + (combined_price * 0.0925)
    return round(total_price, 2)

# Apply the function to create the 'total_price' column
updated_wholesales_df['total_price'] = updated_wholesales_df.apply(lambda row: calculate_total_price(row), axis=1)

# Display the first few rows to check the new column
updated_wholesales_df.head()


```

```python
# Define a function to calculate the cost of each line
def calculate_line_cost(row):
    # Calculate the line cost using unit cost, qty, and shipping price
    line_cost = (row['unit_cost'] * row['qty']) + row['shipping_price']
    return round(line_cost, 2)

# Apply the function to create the 'line_cost' column
updated_wholesales_df['line_cost'] = updated_wholesales_df.apply(lambda row: calculate_line_cost(row), axis=1)

# Display the first few rows to check the new column
updated_wholesales_df.head()



```

```python
# Create a column for the profit of each line using line cost and line price

# Make a copy of the updated DataFrame for the profit calculation
profit_wholesales_df = updated_wholesales_df.copy()

# Create a new column 'profit' that calculates the profit for each line
profit_wholesales_df['profit'] = profit_wholesales_df['total_price'] - profit_wholesales_df['line_cost']

# Display the first few rows to check the new 'profit' column
profit_wholesales_df.head()

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
print("Top 5 client IDs:", top_five_client_ids)

```

```python
# Calculate the total revenue from each of the top 5 clients in Part 1, including first and last names
top_five_clients_revenue = profit_wholesales_df[profit_wholesales_df['client_id'].isin(top_five_client_ids)]\
    .groupby(['client_id', 'first', 'last'])['total_price'].sum().reset_index()

# Sort the DataFrame by 'total_price' in descending order
top_five_clients_revenue = top_five_clients_revenue.sort_values(by='total_price', ascending=False)

# Format the 'total_price' column to two decimal places for better readability
top_five_clients_revenue['total_price'] = top_five_clients_revenue['total_price'].apply(lambda x: f"${x:,.2f}")

# Print the sorted result
print(top_five_clients_revenue)
```

```python
# Calculate the total units, total shipping price, total revenue, and total profit for each of the top 5 clients
summary_df = profit_wholesales_df[profit_wholesales_df['client_id'].isin(top_five_client_ids)]\
    .groupby(['client_id', 'first', 'last']).agg(
        total_units=('qty', 'sum'),
        total_shipping_price=('shipping_price', 'sum'),
        total_revenue=('total_price', 'sum'),
        total_profit=('profit', 'sum')
    ).reset_index()

# Sort the DataFrame by total revenue in descending order
summary_df = summary_df.sort_values(by='total_revenue', ascending=False)

# Display the summary DataFrame
summary_df
```

```python
# Define a function to convert values to millions and format them as strings
def to_millions(value):
    return f"${value / 1_000_000:.2f}M"

# Apply the function to format the summary DataFrame
def format_summary_df(df):
    # Apply the currency formatting to relevant columns
    df['Total Units Purchased'] = df['total_units']
    df['Total Shipping Price (M)'] = df['total_shipping_price'].apply(to_millions)
    df['Total Revenue (M)'] = df['total_revenue'].apply(to_millions)
    df['Total Profit (M)'] = df['total_profit'].apply(to_millions)
    
    # Drop the old columns that are not formatted
    df = df.drop(columns=['total_units', 'total_shipping_price', 'total_revenue', 'total_profit'])
    
    # Sort the DataFrame by 'Total Profit (M)' in descending order
    df = df.sort_values(by='Total Profit (M)', ascending=False)
    
    # Reset the index for clarity
    df = df.reset_index(drop=True)
    
    return df

# Apply the function to the summary DataFrame
formatted_summary_df = format_summary_df(summary_df)

# Display the formatted summary DataFrame
formatted_summary_df
```

```python


# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Bar chart for total revenue
plt.figure(figsize=(10, 6))
sns.barplot(x='first', y='total_revenue', data=summary_df, palette='viridis')
plt.title('Total Revenue by Client (in Millions)')
plt.xlabel('Client')
plt.ylabel('Total Revenue (in Millions)')
plt.show()

# Bar chart for total profit
plt.figure(figsize=(10, 6))
sns.barplot(x='first', y='total_profit', data=summary_df, palette='viridis')
plt.title('Total Profit by Client (in Millions)')
plt.xlabel('Client')
plt.ylabel('Total Profit (in Millions)')
plt.show()

# Stacked bar chart for total units purchased vs. shipping costs
fig, ax = plt.subplots(figsize=(10, 6))
summary_df.set_index('first')[['total_units', 'total_shipping_price']].plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])
plt.title('Total Units Purchased vs. Total Shipping Price')
plt.xlabel('Client')
plt.ylabel('Value')
plt.show()

# Scatter plot of total revenue vs. total profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_revenue', y='total_profit', data=summary_df, hue='first', palette='viridis', s=100)
plt.title('Revenue vs. Profit by Client')
plt.xlabel('Total Revenue (in Millions)')
plt.ylabel('Total Profit (in Millions)')
plt.show()
```
