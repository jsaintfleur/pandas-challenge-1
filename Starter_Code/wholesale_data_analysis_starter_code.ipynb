## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

# Make a copy of the original dataframe to preserve the original data
updated_wholesales_df = wholesales_df.copy()

# Create a new column 'subtotal' that calculates the subtotal for each line
updated_wholesales_df['subtotal'] = updated_wholesales_df['unit_price'] * updated_wholesales_df['qty']

# Display the first few rows to check the new column
updated_wholesales_df.head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.


# Create a new column 'shipping_price' based on the condition of unit_weight
updated_wholesales_df['shipping_price'] = np.where(
    updated_wholesales_df['unit_weight'] > 50,  # Condition: if unit_weight is greater than 50
    updated_wholesales_df['unit_weight'] * 7,    # Shipping price is $7 per pound if the weight is greater than 50
    updated_wholesales_df['unit_weight'] * 10    # Shipping price is $10 per pound if the weight is 50 or under
)

# Display the first few rows to check the new 'shipping_price' column
updated_wholesales_df.head()

```

```python
# Check if there are any items with a unit weight greater than 50 lbs
items_over_50_lbs = updated_wholesales_df['unit_weight'] > 50

# Display the result (True if there are any, False if not)
items_over_50_lbs.any()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

# Define the sales tax rate (convert percentage to a decimal)
sales_tax_rate = 9.25 / 100

# Create a new column 'total_price' that includes subtotal, shipping, and applies sales tax to the sum of both
updated_wholesales_df['total_price'] = (updated_wholesales_df['subtotal'] + updated_wholesales_df['shipping_price']) * (1 + sales_tax_rate)

# Display the first few rows to check the new 'total_price' column
updated_wholesales_df.head()

```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).


# Create a new column 'line_cost' that calculates the total cost for each line
updated_wholesales_df['line_cost'] = (updated_wholesales_df['unit_cost'] * updated_wholesales_df['qty']) + updated_wholesales_df['shipping_price']

# Display the first few rows to check the new 'line_cost' column
updated_wholesales_df.head()


```

```python
updated_wholesales_df.dtypes
```

```python
updated_wholesales_df['order_date'] = pd.to_datetime(updated_wholesales_df['order_date'])
```

```python
updated_wholesales_df.dtypes
```

```python
# Create a column for the profit of each line using line cost and line price

# Make a copy of the updated DataFrame for the profit calculation
profit_wholesales_df = updated_wholesales_df.copy()

# Create a new column 'profit' that calculates the profit for each line
profit_wholesales_df['profit'] = profit_wholesales_df['total_price'] - profit_wholesales_df['line_cost']

# Display the first few rows to check the new 'profit' column
profit_wholesales_df.head()

```

## Part 3: Confirm your work

```python
# Group the DataFrame by 'order_id' and sum the 'total_price' for each order
grouped_totals = profit_wholesales_df.groupby('order_id')['total_price'].sum().reset_index()

# Store the expected totals for the 3 orders from the receipts
expected_totals = {
    2742071: 152811.89,
    2173913: 162388.71,
    6128929: 923441.25
}

# Loop through each order ID and compare the calculated total price with the expected total
for order_id, expected_total in expected_totals.items():
    # Get the calculated total from the grouped data
    calculated_total = grouped_totals[grouped_totals['order_id'] == order_id]['total_price'].values[0]
    
    # Print the result of the comparison
    print(f"Order ID: {order_id}")
    print(f"Calculated Total Price: {calculated_total:.2f}")
    print(f"Expected Total Price: {expected_total:.2f}")
    
    # Check if the calculated total matches the expected total
    if round(calculated_total, 2) == expected_total:
        print(f"Total price for Order ID {order_id} matches the expected total.\n")
    else:
        print(f"Total price for Order ID {order_id} does NOT match the expected total.\n")
```

```python
# Loop through each order ID and break down the components of total_price
for order_id, expected_total in expected_totals.items():
    # Filter the DataFrame for the given order ID
    order_df = profit_wholesales_df[profit_wholesales_df['order_id'] == order_id]
    
    # Sum the subtotal and shipping_price for the order
    subtotal_sum = order_df['subtotal'].sum()
    shipping_sum = order_df['shipping_price'].sum()
    
    # Recalculate the total price by applying sales tax on subtotal + shipping
    sales_tax_rate = 9.25 / 100
    recalculated_total_price = (subtotal_sum + shipping_sum) * (1 + sales_tax_rate)
    
    # Get the actual total price from the DataFrame
    actual_total_price = order_df['total_price'].sum()
    
    # Display the components and the recalculated total
    print(f"Order ID: {order_id}")
    print(f"Subtotal Sum: {subtotal_sum:.2f}")
    print(f"Shipping Price Sum: {shipping_sum:.2f}")
    print(f"Recalculated Total Price: {recalculated_total_price:.2f}")
    print(f"Actual Total Price in Data: {actual_total_price:.2f}")
    print(f"Expected Total Price: {expected_total:.2f}")
    
    # Check if the recalculated total matches the expected total
    if round(recalculated_total_price, 2) == expected_total:
        print(f"Recalculated total price for Order ID {order_id} matches the expected total.\n")
    else:
        print(f"Recalculated total price for Order ID {order_id} does NOT match the expected total.\n")
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Step 1: Find the Top 5 Clients by Quantity
top_5_clients_by_qty = profit_wholesales_df.groupby('client_id')['qty'].sum().nlargest(5).reset_index()

# Step 2: Calculate the Total Spending for Each Client
top_5_clients_spending = profit_wholesales_df.groupby('client_id')['total_price'].sum().reset_index()

# Step 3: Filter to show only the top 5 clients by quantity
top_5_clients_spending = top_5_clients_spending[top_5_clients_spending['client_id'].isin(top_5_clients_by_qty['client_id'])]

# Step 4: Merge total quantity with the spending data
top_5_clients_spending = top_5_clients_spending.merge(top_5_clients_by_qty, on='client_id')

# Step 5: Merge with the original DataFrame to add client names
top_5_clients_spending = top_5_clients_spending.merge(profit_wholesales_df[['client_id', 'first', 'last']].drop_duplicates(), on='client_id')

# Step 6: Format the total_price as currency
top_5_clients_spending['total_price'] = top_5_clients_spending['total_price'].apply(lambda x: f"${x:,.2f}")

# Step 7: Display the results with client names and total quantity
print("Top 5 Clients by Quantity and Their Total Spending:")
print(top_5_clients_spending[['first', 'last', 'client_id', 'qty', 'total_price']])




```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# Step 1: Find the Top 5 Clients by Total Units Purchased (Quantity)
top_5_clients_by_qty = profit_wholesales_df.groupby('client_id')['qty'].sum().nlargest(5).reset_index()

# Step 2: Aggregate the required metrics for the top 5 clients
summary_df = profit_wholesales_df.groupby('client_id').agg({
    'qty': 'sum',                    # Total units purchased
    'shipping_price': 'sum',          # Total shipping price
    'total_price': 'sum',             # Total revenue
    'profit': 'sum'                   # Total profit
}).reset_index()

# Step 3: Filter the summary_df to only include the top 5 clients by quantity
summary_df = summary_df[summary_df['client_id'].isin(top_5_clients_by_qty['client_id'])]

# Step 4: Merge with client names from the original DataFrame
summary_df = summary_df.merge(profit_wholesales_df[['client_id', 'first', 'last']].drop_duplicates(), on='client_id')

# Step 5: Format the total_price and profit as currency
summary_df['total_price'] = summary_df['total_price'].apply(lambda x: f"${x:,.2f}")
summary_df['profit'] = summary_df['profit'].apply(lambda x: f"${x:,.2f}")

# Step 6: Display the final summary DataFrame
print("Summary for the Top 5 Clients:")
summary_df[['first', 'last', 'client_id', 'qty', 'shipping_price', 'total_price', 'profit']]
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 


# Define a function that converts a dollar amount to millions.


# Apply the currency_format_millions function to only the money columns. 


# Rename the columns to reflect the change in the money format. 

# Step 1: Define a function to convert dollar amounts to millions
def currency_format_millions(amount):
    return f"${amount / 1_000_000:.2f}M"

# Step 2: Identify the money columns (total_price and profit)
money_columns = ['total_price', 'profit']

# Step 3: Ensure the columns are numeric and apply the currency_format_millions function
for col in money_columns:
    summary_df[col] = pd.to_numeric(summary_df[col], errors='coerce')  # Convert to numeric if needed
    summary_df[col] = summary_df[col].apply(currency_format_millions)

# Step 4: Rename the columns for presentation
summary_df.rename(columns={
    'qty': 'Total Units Purchased',
    'shipping_price': 'Total Shipping Price',
    'total_price': 'Total Revenue (in millions)',
    'profit': 'Total Profit (in millions)'
}, inplace=True)

# Step 5: Display the formatted DataFrame with renamed columns
print("Formatted Summary for the Top 5 Clients:")
summary_df[['first', 'last', 'client_id', 'Total Units Purchased', 'Total Shipping Price', 'Total Revenue (in millions)', 'Total Profit (in millions)']]







```
